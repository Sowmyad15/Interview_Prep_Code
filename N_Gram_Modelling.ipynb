```python
from nltk.corpus import brown
from nltk.tokenize import word_tokenize
```

```python
corpus = brown.words()
```

```python
corpus
```

```python
corpus_lower=[w.lower() for w in corpus]
vocab = set(corpus_lower)
```

```python
len(corpus_lower)
```

```python
len(vocab)
```

### Find Bigram and Trigram

```python
bigram_count ={}
trigram_count={}
for i in range(len(corpus_lower)-2):
  bigram_words = (corpus_lower[i],corpus_lower[i+1])
  trigram_words = (corpus_lower[i],corpus_lower[i+1],corpus_lower[i+2])

  if bigram_words in bigram_count.keys():
    bigram_count[bigram_words] +=1
  else:
    bigram_count[bigram_words] =1
  if trigram_words in trigram_count.keys():
    trigram_count[trigram_words] +=1
  else:
    trigram_count[trigram_words] =1
```

### Predict Next word

```python
import nltk
nltk.download('punkt_tab')
```

```python
def get_next_word(input,bigram_count,trigram_count,vocab):
  input_words = word_tokenize(input.lower())
  bigram_input = input_words[-2:]

  full_prob={}
  for word in vocab:
    bigram_input_vocab = (bigram_input[0],bigram_input[1])
    trigram_input_vocab = (bigram_input[0],bigram_input[1],word)


    count_bigram_vocab =  bigram_count.get(bigram_input_vocab,0)
    count_trigram_vocab = trigram_count.get(trigram_input_vocab,0)

    if count_bigram_vocab == 0:
      prob = 0
    else:
      prob = count_trigram_vocab/count_bigram_vocab
    full_prob[word] = prob

  full_prob = sorted(full_prob.items(),key=lambda x:x[1],reverse=True)[:3]
  return full_prob
```

```python
get_next_word("I am the king of",bigram_count=bigram_count,trigram_count=trigram_count,vocab=vocab)
```
